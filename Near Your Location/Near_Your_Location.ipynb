```python
import numpy as np
import pandas as pd
import haversine as hs
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
df_camping = pd.read_csv('/content/drive/MyDrive/campingkng.csv')
df_camping = df_camping[['nama_camping', 'kota', 'rating', 'latitude', 'longitude']]
```

```python
df_camping['coor'] = list(zip(df_camping.latitude, df_camping.longitude))
```

```python
# merging latitude and longitude to create a location as accepted by haversine function
df_camping.head()
```

```python
# Function to calculate distance between two locations (user and camping)
# loc_user = location of the user
# loc_camping = location of a camping

import geocoder
g = geocoder.ip('me')

def distance_from(loc_user, loc_camping):
  distance = hs.haversine(loc_user, loc_camping)
  return distance

user_coor = (-6.981039109912842, 108.4928206815238)

distances_km = []
for row in df_camping.itertuples(index=False):
  distances_km.append(distance_from(user_coor, row.coor))

df_camping['distance_from_user'] = distances_km
```

```python
df_camping.sort_values(['distance_from_user'], ascending=[1])
```
