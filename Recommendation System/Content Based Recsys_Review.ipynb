## Content-based Recommendation System

```python
import pandas as pd
```

```python
camping_df = pd.read_csv('campingcuy_fix.csv')
camping_df.head()
```

```python
# check for columns that contain nulls
camping_df.isnull().sum()
```

```python
review_df = pd.read_csv('review_camping.csv')
review_df.head()
```

```python
review_df.isna().sum()
```

```python
review_df.dropna(inplace=True)
```

```python
# group review data for each hotel
review_df = review_df.groupby('camping_id', as_index=True).agg({ 'rating':['mean','count']})
review_df.head()
```

```python
camping_df.head()
```

```python
# merge data camping and review
camping_review_df = camping_df.merge(review_df, how='inner', left_on='camping_id', right_index=True)
```

```python
camping_review_df.head()
```

```python
camping_review_df.columns
```

```python
# rename colums camping_review
camping_review_df.rename(columns={camping_review_df.columns[14]:'rating_mean',
                                  camping_review_df.columns[15]:'ratinguser_count'}, inplace=True)
```

```python
# categorize camping based on price
camping_review_df['harga'] = camping_review_df['harga'].astype(int)
camping_review_df['harga_bin'] = pd.cut(camping_review_df['harga'],
                                      bins=3, labels=['low','medium','high'])
```

```python
camping_review_df.head()
```

```python
# mix column province, city, price_bin as soup metadata
camping_review_df['metadata'] = camping_review_df[['provinsi','kota', 'harga_bin']].agg(' '.join, axis=1)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# calculate TF-IDF and cosine similarity
camping_review_df.set_index('nama_camping', inplace=True)
tf = TfidfVectorizer(analyzer='word', ngram_range=(1, 3), min_df=0,)
tfidf_matrix = tf.fit_transform(camping_review_df['metadata'])
cos_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
cos_sim
```

```python
indices = pd.Series(camping_review_df.index)
indices[:15]
```

```python
def new_recommendations(name, top = 10):

    recommended_camping = []

    idx = indices[indices == name].index[0]
    score_series = pd.Series(cos_sim[idx]).sort_values(ascending = False)

    top = top + 1
    top_indexes = list(score_series.iloc[0:top].index)

    for i in top_indexes:
        recommended_camping.append(list(camping_review_df.index)[i]+" - "+str(score_series[i]))

    return recommended_camping
```

```python
# example of display camping recommendations
new_recommendations('Ranca Upas Camping Ground', 15)
```
